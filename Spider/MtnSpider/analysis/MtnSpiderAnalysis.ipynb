Next, lets create Histograms for  three key numerical variables:

Climb Heights (ft):

The distribution is heavily skewed right, with most climbs being relatively short, as seen by the concentration of data near lower heights. There's a long tail, indicating some climbs are significantly taller.

Page Views per Month:

This distribution is also right-skewed, with most climbs receiving fewer than 20 views per month, but a few climbs receive a disproportionately high number of views.

Average Stars:

The ratings distribution appears multi-modal, with peaks around whole and half-star ratings (2, 2.5, 3), suggesting climbers prefer to rate climbs at these intervals.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the figure and axes for the histograms
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Histogram for Climb Height in feet
sns.histplot(df['Climb Height (ft)'].dropna(), bins=50, ax=axes[0], kde=True)
axes[0].set_title('Histogram of Climb Heights (ft)')
axes[0].set_xlabel('Climb Height (ft)')
axes[0].set_ylabel('Frequency')

# Histogram for Page Views per Month
sns.histplot(df['Page Views per Month'], bins=50, ax=axes[1], kde=True)
axes[1].set_title('Histogram of Page Views per Month')
axes[1].set_xlabel('Page Views per Month')
axes[1].set_ylabel('Frequency')

# Histogram for Average Stars
sns.histplot(df['Average Stars'].dropna(), bins=50, ax=axes[2], kde=True)
axes[2].set_title('Histogram of Average Stars')
axes[2].set_xlabel('Average Stars')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Next, let's create box plots for the same variables to identify outliers and visualize the spread of the data more clearly.


```python
# Setting up the figure and axes for the box plots
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Box plot for Climb Height in feet
sns.boxplot(x=df['Climb Height (m)'], ax=axes[0])
axes[0].set_title('Box Plot of Climb Heights (m)')
axes[0].set_xlabel('Climb Height (m)')

# Box plot for Page Views per Month
sns.boxplot(x=df['Page Views per Month'], ax=axes[1])
axes[1].set_title('Box Plot of Page Views per Month')
axes[1].set_xlabel('Page Views per Month')

# Box plot for Average Stars
sns.boxplot(x=df['Average Stars'], ax=axes[2])
axes[2].set_title('Box Plot of Average Stars')
axes[2].set_xlabel('Average Stars')

plt.tight_layout()
plt.show()
```
