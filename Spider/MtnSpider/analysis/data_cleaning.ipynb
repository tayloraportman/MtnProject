```python
import pandas as pd 

df = pd.read_csv('combined_data_all.csv')
df.head(5)
```


Here are the percentages of missing values in the dataset, sorted by the highest missing rate:

Next steps:
remove Grade Font -- remaining analysis will be in YDS
review number of pitches -- since high missing rate, drop num of pitches

```python
# drop Grade Font -- remaining analysis will be in YDS
data_clean = df.drop(columns=['Grade Font'])
# Drop rows where 'Grade YDS' is missing
data_clean = data_clean.dropna(subset=['Grade YDS'])
```

```python
# remove data with Climb Height (m) > 30,000m
data_clean = data_clean[data_clean['Climb Height (m)'] <= 1500]
```

```python
# save cleaned data to csv
data_clean.to_csv('cleaned_dat.csv', index=False)
```
